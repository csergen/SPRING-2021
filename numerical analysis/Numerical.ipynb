```python
import numpy as np
import pandas as pd
from sympy import *

x, y, z = symbols('x y z')

class Nume:

    @staticmethod
    def cramer(A: np.ndarray, b: np.ndarray) -> np.ndarray:
        if np.shape(A) != (len(A[0, :]), len(A[:, 0])):
            print(f"A is not a square matrix: A{ np.shape(A) }")
            return np.NaN
    
        delta_n = []
    
        for i in range(len(b)):
            copy_A = A.copy()
            copy_A[:, i] = b
        
            resval = np.linalg.det(copy_A) / np.linalg.det(A)
            delta_n.append(resval)
    
        return np.array(delta_n)

    @staticmethod
    def bisection(equ, interval: list, itr: int,
                  epsilon: float=None,
                  real_root: float=None) -> pd.DataFrame:
        err: float = None
        arr: list = []
        root: float = lambda t: float(equ.subs(x, t))
        chrt: bool = lambda a, b: ( root(a) * root(b) < 0 )
        counter: int = 1
        
        while (itr > 0):
            lower, upper = interval
            if chrt(lower, upper):
                aprx = (lower + upper) / 2
                
                if chrt(lower, aprx):
                    interval[1] = aprx
                elif chrt(aprx, upper):
                    interval[0] = aprx
                
                if real_root:
                    err = np.abs(real_root - aprx)
                elif epsilon:
                    err = (upper - lower) / pow(2, counter)
                
                arr.append([lower, upper, aprx, err, epsilon])
                
                if (err and epsilon) and (err < epsilon):
                    break

                itr -= 1
            else:
                break
        
        arr = np.array(arr)
                
        df = pd.DataFrame({"lower": arr[:, 0],
                           "upper": arr[:, 1],
                           "approximate": arr[:, 2],
                           "error": arr[:, 3],
                           "epsilon": arr[:, 4]})
        df.index += 1
        
        return df
    
    @staticmethod
    def check(A: np.ndarray, x: np.ndarray, b: np.ndarray) -> bool:
        return np.allclose(np.dot(A, x), b)
    
    # Upper triangular form
    @staticmethod
    def triu(A: np.ndarray) -> np.ndarray:
        return np.triu(A)
    
    # Lower triangular form
    @staticmethod
    def tril(A: np.ndarray) -> np.ndarray:
        return np.tril(A)
    
    # Diagonal matrix form
    @staticmethod
    def diag(A: np.ndarray) -> np.ndarray:
        return np.diag(np.diag(A))
    
    @staticmethod
    def ispos(a):
        if a > 0:
            return True
        return False
    
    @staticmethod
    def isneg(a):
        if a < 0:
            return True
        return False
    
    @staticmethod
    def iszero(a):
        if a == 0:
            return True
        return False
```

```python
#@title Variables
equ = 'x**3-x-2' #@param {type: "string"}
lower = 1 #@param {type: "number"}
upper = 2 #@param {type: "number"}
iteration = 15 #@param {type: "integer"}
epsilon = 0 #@param {type: "number"}
real_root = 0 #@param {type: "number"}
interval = [lower, upper]

resval = Nume.bisection(eval(equ), interval, iteration, epsilon, real_root)
resval
```
